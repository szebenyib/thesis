## Genetic Algorithm

```python
#Tree representation
from pyevolve import GTree

#Evolution support
from pyevolve import GSimpleGA
from pyevolve import Consts

#Error accumulator
from pyevolve import Util

#Math support for sqrt
import math
```

```python
rmseAccum = Util.ErrorAccumulator()
```

### GP operators

Gp prefix ensures that Pyevolve adds them as non-terminals.

```python
def gp_add(a, b):
    return a + b

def gp_sub(a, b):
    return a - b

def gp_mul(a, b):
    return a * b

def gp_sqrt(a):
    return math.sqrt(abs(a))
```

### Evaluation function

```python
def eval_func(chromosome):
    #Using the global variable
    global rmseAccum
    rmseAccum.reset()
    
    #Avoid parsing by compiling into bytecode
    codeComp = chromosome.getCompiledCode()
    
    for a in xrange(0, 5):
        for b in xrange(0, 5):
            evaluated = eval(codeComp)
            target = math.sqrt((a * a) + (b * b))
            rmseAccum += (target, evaluated)
  
    return rmseAccum.getRMSE()
```

### Step callback

Run every time after the generation has been evaluated. Use it to get progression information.

```python
def step_callback(gp_engine):
    if gp_engine.getCurrentGeneration() % 10 == 0:
        GTree.GTreeGP.writePopulationDot(
            gp_engine,
            "trees_" + str(gp_engine.getCurrentGeneration()) + ".jpg",
            start = 0,
            end = 3)
```

### The evolution

For parallel computing capabilities the evolution has got to be put in a function and later on called by "\__main\__".

```python
def main_run():
    #Individuals' sketch
    genome = GTree.GTreeGP()
    genome.setParams(max_depth = 4, method = "ramped")
    genome.evaluator.set(eval_func)
    
    #Engine
    ga = GSimpleGA.GSimpleGA(genome)
    ga.setParams(gp_terminals = ['a', 'b'], #Terminals, same as in eval_func()
                 gp_function_prefix = "gp") #Non-terminals, operators
    ga.stepCallback.set(step_callback)

    
    #Evolution parameters
    ga.setMinimax(Consts.minimaxType["minimize"])
    ga.setGenerations(30)
    ga.setCrossoverRate(1.0)
    ga.setMutationRate(0.25)
    ga.setPopulationSize(400)
    
    ga(freq_stats = 10)
    print ga.bestIndividual()
```

## Run

```python
if __name__ == "__main__":
    main_run()
```
